### A performance test I wrote to see if I need a warmup script

```python
import time
# warm up (starts JVM & Python worker)
t0 = time.perf_counter()
_ = spark.range(1).count()
warmup = time.perf_counter() - t0
print("warmup secs:", round(warmup, 1))

# faster small-DF creation (no schema infer)
t0 = time.perf_counter()
df_fast = spark.createDataFrame([(1,), (2,), (3,)], schema=["value"])
print("create secs:", round(time.perf_counter() - t0, 1))
t0 = time.perf_counter()
df_fast.show()
show_secs = time.perf_counter() - t0
print("show secs:", round(show_secs, 1))
print("total secs:", round(warmup + show_secs, 1))
```

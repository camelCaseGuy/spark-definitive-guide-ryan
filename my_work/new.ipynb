```python
from spark_init import start_spark

spark = start_spark()
spark
```

```python
myRange = spark.range(1000).toDF("number").where("number > 990")
myRange.count() # This is an action.

myRange.show() # This is also an action.
myRange.describe().show() # This is also an action.
myRange.printSchema() # This is not an action.
myRange.explain() # This is not an action.
myRange.cache() # This is not an action.
myRange.count() # This is an action.
myRange.unpersist() # This is not an action.
myRange.count() # This is an action.
# spark.stop()
```
